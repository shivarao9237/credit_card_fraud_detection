```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_data=pd.read_csv("creditcard.csv")
```

```python
credit_data.head()
```

```python
credit_data.info()
```

```python
#check no of missing values in each col
credit_data.isnull().sum()
```

```python
#distribution of legit and fraud
credit_data['Class'].value_counts()
```

```python
#it is an unblance data set because normal trans are more than legit
#seperating data for analysis
legit=credit_data[credit_data.Class==0]
fraud=credit_data[credit_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare values for both transactions
credit_data.groupby('Class').mean()
```

```python
#build a sample dataset containing similar distribution of normal and faud transaction
legit_sample=legit.sample(n=492)
```

```python
#concatenating two dataframes
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#split data into features and targets
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
```

```python
print(x)
```

```python
y
```

```python
#split the data into train and test data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#training the model
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#accuracy on train data
x_train_pred=model.predict(x_train)
train_accuracy=accuracy_score(x_train_pred,y_train)
print(train_accuracy)
```

```python
#accuracy on test data
x_test_pred=model.predict(x_test)
test_accuracy=accuracy_score(x_test_pred,y_test)
print(test_accuracy)
```
